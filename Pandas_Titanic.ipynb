```python
titanic = pd.read_csv("/content/train.csv")

# Quick look
titanic.head()
```

```python
titanic.info()      # Check columns, types, missing values
titanic.describe()  # Summary statistics
```

```python

# Fill missing Age with median
titanic['Age'] = titanic['Age'].fillna(titanic['Age'].median())

# Fill missing Embarked with mode
titanic['Embarked'] = titanic['Embarked'].fillna(titanic['Embarked'].mode()[0])

# Drop Cabin column if it exists
titanic = titanic.drop(columns=['Cabin'], errors='ignore')

# Remove duplicates
titanic = titanic.drop_duplicates()

# Check changes
titanic.info()
```

```python
# Survival rate by gender
survival_by_gender = titanic.groupby('Sex')['Survived'].mean()
print("Survival rate by gender:\n", survival_by_gender)

# Survival rate by class
survival_by_class = titanic.groupby('Pclass')['Survived'].mean()
print("\nSurvival rate by class:\n", survival_by_class)

# Average age per class
average_age_class = titanic.groupby('Pclass')['Age'].mean()
print("\nAverage age per class:\n", average_age_class)

# Survival rate by age group
# Define age groups
bins = [0, 12, 18, 35, 60, 100]
labels = ['Child', 'Teen', 'Adult', 'Mid-age', 'Senior']
titanic['AgeGroup'] = pd.cut(titanic['Age'], bins=bins, labels=labels)

survival_by_agegroup = titanic.groupby('AgeGroup', observed=False)['Survived'].mean()
print("\nSurvival rate by age group:\n", survival_by_agegroup)
```

```python
# Female passengers who survived
female_survived = titanic[(titanic['Sex'] == 'female') & (titanic['Survived'] == 1)]
print("Female passengers who survived:\n", female_survived[['Name','Sex','Survived']])

# Children who survived (age <= 12)
children_survived = titanic[(titanic['Age'] <= 12) & (titanic['Survived'] == 1)]
print("\nChildren who survived:\n", children_survived[['Name','Age','Survived']])

# 1st class passengers with high survival probability
first_class_survived = titanic[(titanic['Pclass'] == 1) & (titanic['Survived'] == 1)]
print("\n1st class passengers who survived:\n", first_class_survived[['Name','Pclass','Survived']])
```

```python
# Gender impact
most_likely_gender = "female" if survival_by_gender['female'] > survival_by_gender['male'] else "male"
print(f"More likely to survive: {most_likely_gender}")

# Class impact
print("Survival increases with higher class:", survival_by_class)

# Children prioritization
child_survival_rate = survival_by_agegroup['Child']
print("Children survival rate:", child_survival_rate)

# Highest survival combination
# Typically: Female + 1st class
highest_combo = titanic[(titanic['Sex']=='female') & (titanic['Pclass']==1)]
highest_combo_rate = highest_combo['Survived'].mean()
print("Highest survival rate combination (Female & 1st class):", highest_combo_rate)
```

✅ Insights Summary:

Females were more likely to survive.
Higher class passengers had better survival.
Children had relatively high survival rates.
Female in 1st class had the highest survival probability.
